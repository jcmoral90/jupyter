#### B. LaTeX

Puedes usar LaTeX en una línea como $y = x^2$ o en bloque $$y = x^2$$

- Operaciones
    - Suma:
        - $x + y$
    - Resta:
        - $x - y$          
    - Multiplicación
        - $x * y$
    - División
        - $x / y$
- Relaciones
    - $\pi \approx 3.14159$
    - ${1 \over 0} \neq \infty$
    - $0 < x > 1$
    - $0 \leq x \geq 1$
    
- Fracciones
    - $1/_2$
    - $\frac{1}{2x}$
    - ${3 \over 4}$
    
- Letras del alfabeto griego (algunas de ejemplo)
    - $\alpha$
    - $\tau$
    - $\sigma$
    
- Potencias y Subindices
Podemos crear indices y subindices en ecuaciones

    - $$f(x) = x^2 - x^{1 \over \pi}$$
    - $$f(X,n) = X_n + X_{n-1}$$
    
- Raices y logratimos

    - $$ f(x) = \sqrt[3]{2x} + \sqrt{x-2}$$
    - $$\log_x x = 1$$
    
- Sumas y productos

    - $$\sum\limits_{x=1}^{\infty} {1 \over x} = 2$$
    - $$\prod\limits_{i=1}^{n} x_i - 1$$
    
- Estadistica

    - $$max(S) = \max\limits_{i: x_i \in \{S\}} x_i$$
    - $$min (S) = \min\limits_{i: x_i \in \{S\}} x_i$$
    - $$\frac{n!}{k!(n-k)!} = {n \choose k}$$
    
- Calculo
 
    - $$\lim\limits_{x \to 0^+} {1 \over 0} = \inf$$
    - $$\int\limits_a^b 2x \, dx$$
    
- Definiciones de funciones sobre periodos

    - $f(x) =
      \left\{
          \begin{array}{ll}
              0  & \mbox{if } x = 0 \\
              {1 \over x} & \mbox{if } x \neq 0
          \end{array}
      \right.$
